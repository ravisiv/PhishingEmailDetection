```python
import json
from jsonpath_ng import jsonpath, parse
with open('/Users/ravis/Downloads/ex2.metadata.json') as json_file:
    data = json.load(json_file)

jsonpath_expr= parse("$.classes")
app_attributes = []
people_attributes = []

for key in jsonpath_expr.find(data):
    classes  = key.value
    for classes_key, classes_val in enumerate(classes):
        if classes.get(classes_val)["name"] == "Applications":
            attrib = classes.get(classes_val)["attributes"]
            for each_attr in attrib:
                app_dict = {}
                app_dict["name"] = each_attr["name"]
                app_dict["id"] = each_attr["id"]
                app_attributes.append(app_dict)
        elif classes.get(classes_val)["name"] == "People":
            attrib = classes.get(classes_val)["attributes"]
            for each_attr in attrib:
                app_dict = {}
                app_dict["name"] = each_attr["name"]
                app_dict["id"] = each_attr["id"]
                people_attributes.append(app_dict)

    
```

```python
applications_dict = {}
applications_dict["name"] = "Applications"
applications_dict["attributes"] = app_attributes

people_dict = {}
people_dict["name"] = "People"
people_dict["attributes"] = people_attributes

final_dict = {
    
}

final_dict["classes_with_virtual_attributes"] = [applications_dict,people_dict]
```
