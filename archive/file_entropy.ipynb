```python
from scipy.stats import entropy
import pandas as pd
```

```python
def get_entropy_from_file(attachment_file):

    data = bytearray()
    with open(attachment_file, "rb") as f:
        while (byte := f.read(1)):
            data+=bytearray(byte)

    entroy_file = entropy(data,base=2)
       
    return entroy_file
```

```python
rootdir="/Users/ravis/Downloads/NapierOne-tiny"
rows = []
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        filepath = subdir + os.sep + file
        if os.path.isfile(filepath):
            file_ext = os.path.splitext(str(filepath))[1]
            file_root = os.path.splitext(str(filepath))[0]
            file_size = os.path.getsize(str(filepath))
            entroy_file = get_entropy_from_file(str(filepath))
            rows.append([file_root,file_ext,file_size,entroy_file ])
   
```
